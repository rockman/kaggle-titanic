```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('train.csv', index_col='PassengerId')
test_df = pd.read_csv('test.csv', index_col='PassengerId')
```

```python
df = pd.concat([train_df, test_df.assign(Survived=-1)], sort=False)
df.sample(5)
```

```python
df.isna().sum()
```

```python
df.Name.str.match('[^,]+,\s[^.]+\.\s.*').all()
```

```python
df['Title'] = df.Name.str.extract('[^,]+,\s([^.]+)\.\s.*')
```

```python
df[['Name', 'Title']].head()
```

```python
df.Title.unique()
```

```python
df.groupby('Title').Age.median()
```

```python
df.Age = df.groupby('Title').Age.apply(lambda x: x.fillna(x.median()))
df.head()
```

```python
df.isna().sum()
```

```python
df.Fare = df.Fare.fillna(df.Fare.median())
```

```python
df.isna().sum()
```

```python
df.Cabin.unique()
```

```python
df.Cabin = df.Cabin.fillna('Z')
df.Cabin = df.Cabin.str.get(0)
df.Cabin.unique()
```

```python
df.isna().sum()
```

```python
df_nan_embarked = df[df.Embarked.isna()]
df_nan_embarked
```

```python
df[df.Ticket == '113572']
```

```python
df.Fare.describe()
```

```python
std_fare = df.Fare.std()
df[(df.Pclass == 1) & df.Fare.between(80 - std_fare, 80 + std_fare)].Embarked.value_counts()
```

```python
df.Embarked = df.Embarked.fillna('S')
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df = df.drop(columns=['Name', 'Ticket', 'Title'])
```

```python
df.head()
```

```python
df.Sex = df.Sex.map({'male': 1, 'female': 0})
```

```python
df.head()
```

```python
df_with_dummies = pd.get_dummies(df)
df_with_dummies.head()
```

```python
mask = df_with_dummies.Survived == -1
test_df = df_with_dummies[mask]
train_df = df_with_dummies[~mask]
y = train_df[['Survived']]
X = train_df.drop(columns='Survived')
submission_X = test_df.drop(columns='Survived')
```
