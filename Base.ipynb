```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
df = pd.concat([train_df, test_df.assign(Survived=-1)], sort=False)
```

```python
df.sample(5)
```

```python
df = df.set_index('PassengerId')
df.head()
```

```python
df.info()
```

For this base notebook, will keep it simple and drop:
- Name (could be helpful with missing age values)
- Ticket (could families be on the same ticket?)
- Cabin (this is probably very useful, but needs more attention; also, the number of missing values is very high)

```python
df.drop(columns=['Name', 'Ticket', 'Cabin'], inplace=True)
df.head()
```

```python
df.isna().sum()
```

```python
df.Sex = df.Sex.map({'male': 1, 'female': 0})
df.head()
```

Normally, we would impute with more care, but since this notebook is the base, we will just take median/mode/etc

```python
train_subset = df[df.Survived != -1]
```

```python
train_subset.describe(include='all')
```

```python
df.Age.fillna(train_subset.Age.median(), inplace=True)
df.Fare.fillna(train_subset.Fare.median(), inplace=True)
df.Embarked.fillna(train_subset.Embarked.mode()[0], inplace=True)
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df_with_dummies = pd.get_dummies(df)
df_with_dummies.head()
```

```python
mask = df_with_dummies.Survived == -1
test_subset = df_with_dummies[mask]
train_subset = df_with_dummies[~mask]
y = train_subset.Survived
X = train_subset.drop(columns=['Survived'])
submission_X = test_subset.drop(columns=['Survived'])
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
model_logr = LogisticRegression()
model_logr.fit(X_train, y_train)
model_logr.score(X_test, y_test)
```

```python
submission_X.head()
```

```python
submission_predictions = model_logr.predict(submission_X)
```

```python
submission_df = pd.DataFrame({'PassengerId': submission_X.index, 'Survived': submission_predictions})
```

```python
submission_df.head()
```

```python
submission_df.to_csv('base_submission.csv', index=False)
```
