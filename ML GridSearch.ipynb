```python
import pandas as pd
import numpy as np
```

```python
X = pd.read_csv('X.csv')
y = pd.read_csv('y.csv')
submission_X = pd.read_csv('submission_X.csv')
```

```python
X.head()
```

```python
X = X.drop(columns='PassengerId')
```

```python
X.head()
```

```python
submission_X.head()
```

```python
submission_X_passengers = submission_X.PassengerId
submission_X = submission_X.drop(columns='PassengerId')
```

```python
submission_X.head()
```

```python
y.head()
```

```python
y = y.Survived
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],
    'solver': ['lbfgs'],
    'max_iter': [500],
}
```

```python
grid = GridSearchCV(LogisticRegression(), param_grid=params, cv=5, scoring='roc_auc')
```

```python
grid.fit(X_train, y_train)
```

```python
grid.score(X_test, y_test)
```

```python
submission_predictions = grid.predict(submission_X)
```

```python
submission_df = pd.DataFrame({'PassengerId': submission_X_passengers, 'Survived': submission_predictions})
```

```python
submission_df.head()
```

```python
submission_df.to_csv('ml_gridsearch_submission.csv', index=False)
```
